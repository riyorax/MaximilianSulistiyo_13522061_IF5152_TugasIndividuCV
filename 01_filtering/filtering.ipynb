# 1 Import Libraries and Images

Import Library

```python
import cv2
import matplotlib.pyplot as plt
```

Import images

```python
from skimage import data 
cameraman = data.camera()
coin = data.coins()
checkerboard = data.checkerboard()
maisan = cv2.imread("../test_pictures/maisan.png", cv2.IMREAD_GRAYSCALE)
seiko = cv2.imread("../test_pictures/seiko_panda.jpg", cv2.IMREAD_GRAYSCALE)
```

# 2 Filter Functions

## Gaussian Filter

```python
def apply_gaussian_filter(image, kernel_size, sigma):
    """
    Apply Gaussian filter to the image
    
    Args:
        image: Input image (grayscale or color)
        kernel_size: Size of the Gaussian kernel (must be odd)
        sigma: Standard deviation of Gaussian kernel
    
    Returns:
        Filtered image
    """
    filtered = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
    
    return filtered
```

## Median Filter

```python
def apply_median_filter(image, kernel_size):
    """
    Apply Median filter to the image
    
    Args:
        image: Input image (grayscale or color)
        kernel_size: Size of the median kernel (must be odd)
    
    Returns:
        Filtered image
    """
    filtered = cv2.medianBlur(image, kernel_size)
    
    return filtered
```

## Testing Pipeline

```python
def show_image(image, image_name):
    plt.imshow(image, cmap='gray')
    plt.title(f'{image_name}')
    plt.axis('off')
    plt.show()
```

```python
def test_filter(image, image_name, kernel_size_gaussian, kernel_size_median, sigma):
    # Show Original Image
    show_image(image, image_name)

    # Filter the Images
    image_gaussian_filtered = apply_gaussian_filter(image, kernel_size_gaussian, sigma)
    image_median_filtered = apply_median_filter(image, kernel_size_median)

    # Show filtered images
    show_image(image_gaussian_filtered, f"{image_name} Gaussian Filter")
    show_image(image_median_filtered, f"{image_name} Median Filter")

    # Save filtered images
    cv2.imwrite(f"{image_name}_gaussian.png", image_gaussian_filtered)
    cv2.imwrite(f"{image_name}_median.png", image_median_filtered)
```

# 3 Testing

### 3.1 Cameraman

```python
test_filter(cameraman, "Cameraman", kernel_size_gaussian=15, kernel_size_median=7, sigma=5)
```

### 3.2 Coin

```python
test_filter(coin, "Coin", kernel_size_gaussian=5, kernel_size_median=15, sigma=3)
```
